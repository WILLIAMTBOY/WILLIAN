```python
# DataFrame with the given data
data = {
    'CENSUS2010POP': [21720, 2756, 23796, 19198, 9315, 195408, 18021, 160331, 14755, 7208],
    'STATENAME': ['Texas', 'Georgia', 'Virginia', 'Kentucky', 'Missouri', 'Texas', 'Georgia', 'Virginia', 'Kentucky', 'Missouri']
}

df = pd.DataFrame(data)

# Group by state and sum the population for each state
state_populations = df.groupby('STATENAME')['CENSUS2010POP'].sum()

# Sort the states by population in descending order
sorted_states = state_populations.sort_values(ascending=False)

#  top three most populous states
top_three_states = sorted_states.head(3)

print("The three most populous states (in order of highest population to lowest population) are:")
for i, (state, population) in enumerate(top_three_states.iteritems(), start=1):
    print(f"{i}. {state}: {population} people")
```
